# Transforming Skewed Continuous Features

A dataset may sometimes contain at least one feature whose values tend to lie near a single number, but will also have a non-trivial number of vastly larger or smaller values than that single number. Algorithms can be sensitive to such distributions of values and can underperform if the range is not properly normalized.

```python
# check features for skewness using histogram
vs.histogram(features_train)
```

```python
# funtion for log transformation of skewed features
def skew(df,skewed):
    features_log_transformed = pd.DataFrame(data=df)
    features_log_transformed[skewed] = df[skewed].apply(lambda x:np.log(x+1))
    return features_log_transformed

skewed = ['Open','Volume','P/E','Div Yield','Vix','O-H','H-L','O-L']


features_train_log_transformed = skew(features_train,skewed)
features_test_log_transformed = skew(features_test,skewed)
#vs.histogram(features_train_log_transformed)
#display(features_train_log_transformed.head())
```

# Normalizing Numerical Features

In addition to performing transformations on features that are highly skewed, it is often good practice to perform some type of scaling on numerical features. Applying a scaling to the data does not change the shape of each feature's distribution (such as 'capital-gain' or 'capital-loss' above); however, normalization ensures that each feature is treated equally when applying supervised learners. Note that once scaling is applied, observing the data in its raw form will no longer have the same original meaning


```python
# Function for normalizing and denormalizing of data using df/df[1] method
def normalize(data,row):
    data = data/row
    return data

def denormalize(data,row):
    data = data*row
    return data

feature_scale = features_train_log_transformed.iloc[0,:]
prices_scale = prices.iloc[0]
features_train_normal = normalize(features_train_log_transformed,feature_scale)
features_test_normal = normalize(features_test_log_transformed,feature_scale)
prices_train_normal = normalize(prices,prices_scale)
prices_test_normal = normalize(prices_test,prices_scale)
#display(features_train_normal.head())
#display(features_test_normal.head())
#display(prices_train_normal.head())
#display(prices_test_normal.head())
#type(prices_test)
```

# Evaluating Model Performance

It is difficult to measure the quality of a given model without quantifying its performance over training and testing. This is typically done using some type of performance metric, whether it is through calculating some type of error, the goodness of fit, or some other useful measurement. For this project, we will be calculating the RMSE error and coefficient of determination, R2, to quantify model's performance. 

```python
from sklearn.metrics import r2_score, mean_squared_error

# function for model evaluation using RMSE error and R2 score
def performance_metric(y_true,y_pred):
    score = r2_score(y_true,y_pred)
    error = mean_squared_error(y_true,y_pred)
    return np.sqrt(error),score
```

# Creating a benchmark model

## Benchmark Model : Linear Regression

```python
from sklearn.linear_model import LinearRegression
global SEED
SEED = 42

# function for training and predicting LinearRegression Model
def lr_predict(base_model):
    base_model.fit(X_train,y_train)
    lrtest = base_model.predict(X_test)
    lrtrain = base_model.predict(X_train)
    return lrtrain, lrtest

# function for plotting actual and predicted prices
def model_plot(df,title):    
    
    df1 = pd.DataFrame(data=prices_test)
    df1['Pred'] = df
    fig = plt.figure(figsize=(28,12))
    plt.plot(df1['Close'],color='black')
    plt.plot(df1['Pred'],color='blue')
    plt.xlabel('Date',fontsize=(20))
    plt.ylabel('Prices',fontsize=(20))
    plt.title(title,fontsize=(30))
    plt.legend(["Actual","Prediction"],fontsize=(20))
    plt.show()
```

```python
# Implement Linear Regression Model
lrmodel = LinearRegression()
lrtrain,lrtest = lr_predict(lrmodel)
lrtestden = denormalize(lrtest,prices_scale)
rmse,score = performance_metric(prices_test,lrtestden)
print("{} Model has RMSE Testing error of {:.2f}".format(lrmodel.__class__.__name__,rmse))
print("{} Model has R2 score of {:.2f}".format(lrmodel.__class__.__name__,score))
model_plot(lrtestden,'Linear Regression Model Prediction')
```

# Model 1: Stacking of Learners

```python
from sklearn.svm import LinearSVR
from sklearn.tree import DecisionTreeRegressor
from keras.models import Sequential
from keras.layers.core import Dense,Activation,Dropout
from keras.layers.recurrent import LSTM
from xgboost import XGBRegressor
from sklearn.ensemble import  GradientBoostingRegressor
from sklearn.ensemble import  RandomForestRegressor
from sklearn.ensemble import AdaBoostRegressor
from sklearn.ensemble import ExtraTreesRegressor
from sklearn.model_selection import TimeSeriesSplit
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import make_scorer
```

```python
# train and predict models using GridSearchCV

def base_model_predict(model,params):
    scorer = make_scorer(mean_squared_error)
    tscv = TimeSeriesSplit(n_splits=5)
    tscv = tscv.split(range(X_train.shape[0]))
    grid_obj = GridSearchCV(model,params,scorer,cv=tscv)
    grid_fit = grid_obj.fit(X_train,y_train)
    best_reg = grid_fit.best_estimator_
    ypred = best_reg.predict(X_test)
    ytrain = best_reg.predict(X_train)
    return best_reg, ytrain, ypred
```

### Base Model 1: AdaBoost Regressor

```python
adareginit = AdaBoostRegressor(random_state=SEED)
adaparamsinit = {}
adaregmodelinit,adatraininit,adatestinit = base_model_predict(adareginit,adaparamsinit)
adatestdeninit = denormalize(adatestinit,prices_scale)
adaerrorinit,adascoreinit = performance_metric(prices_test,adatestdeninit)
print("{} has Unoptimized Testing RMSE error of {:.2f}".format(adareginit.__class__.__name__,adaerrorinit))
print("{} has Unoptimized R2 score of {:.2f}".format(adareginit.__class__.__name__,adascoreinit))
```

```python
adareg = AdaBoostRegressor(base_estimator=DecisionTreeRegressor(),random_state=42)
# parameters already tuned using Gridsearchcv
adaparams = {'n_estimators':[50],'learning_rate':[0.01],'base_estimator__max_depth':[10],'base_estimator__min_samples_split':[7]}
adaregmodel,adatrain,adatest = base_model_predict(adareg,adaparams)
adatestden = denormalize(adatest,prices_scale)
adaerror,adascore = performance_metric(prices_test,adatestden)
print("{} has Testing RMSE error of {:.2f}".format(adareg.__class__.__name__,adaerror))
print("{} has R2 score of {:.2f}".format(adareg.__class__.__name__,adascore))
adaooftrain = adatrain.reshape(-1,1)
adaooftest = adatest.reshape(-1,1)
model_plot(adatestden,'AdaBoost Regressor Prediction')
```

# Stacker Model: XGB Regressor

```python
def xgb_model_predict(model,params):
    scorer = make_scorer(mean_squared_error)
    tscv = TimeSeriesSplit(n_splits=5)
    tscv = tscv.split(range(X_train.shape[0]))
    grid_obj = GridSearchCV(model,params,scorer,cv=tscv)
    grid_fit = grid_obj.fit(x_train,y_train)
    best_reg = grid_fit.best_estimator_
    ypred = best_reg.predict(x_test[:200,])
    ytrain = best_reg.predict(x_train)
    return best_reg, ytrain, ypred
```

```python
xgbreg = XGBRegressor()
# parameters already tuned using Gridsearchcv
xgbparams = {'n_estimators':[100],'learning_rate':[0.07],
             'gamma':[0.0009],'max_depth':[3],'random_state':[SEED]}
xgbregmodel,xgbtrain,xgbtest = xgb_model_predict(xgbreg,xgbparams)
xgbtestden = denormalize(xgbtest,prices_scale)
xgberror,xgbscore = performance_metric(prices_test[:200,],xgbtestden)
print("{} has Testing error of {:.2f}".format(xgbreg.__class__.__name__,xgberror))
print("{} has R2 score of {:.2f}".format(xgbreg.__class__.__name__,xgbscore))
model_plot(xgbtestden, 'XGB Regressor Predictions')
```

As can be seen from above plot, predictions very closely resembles the actual price trend.
